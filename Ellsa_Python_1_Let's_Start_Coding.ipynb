# Multiple Code

Satu cell dapat menampung lebih dari satu baris code. Cobalah untuk mencetak beberapa kalimat sekaligus.

multiplecode.PNG

```python
print ("Halo  Dunia")
print ('Selamat pagi!')
print ("Phyton sangat mudah untuk dipelajari dan digunakan")
```

Python menjalankan/mengeksekusi code baris perbaris, dimulai dari baris yang paling atas hingga baris paling bawah.
Hal ini bisa kita lihat pada output code sebelumnya.
Output yang tampil dimulai dari "*Halo Dunia*" dan diakhiri "*Python sangat mudah untuk dipelajari dan digunakan*"


multiplecoderun.PNG


Kalau kita amati, output code sebelumnya teridiri dari 3 baris kalimat. Padahal, kita tidak menambahkan *enter* di akhir setiap kalimat.

Hal ini karena Python secara otomatis menambahkan baris baru (*enter*) di akhir perintah <code>print()</code>.

Jika kita tidak ingin Python menambahkan baris baru diakhir kalimat, maka tambahkan <code>end</code> pada perintah <code>print()</code>.


end.PNG

```python
print('Sambung ', end='')
print('menyambung ', end='')
print('menjadi ', end='')
print('satu', end='')
```

Sejauh ini, kita hanya menggunakan `print()` untuk mencetak satu kalimat. Bila kita ingin mencetak lebih dari satu kalimat dengan banyak baris, maka ubah tanda petik tunggal atau ganda menjadi triple quotes (`docstrings`).

# Escape Character

Kita sudah menyinggung escape character sebelumnya.
Escape character (`\`) ini juga dapat kita kombinasikan dengan karakter khusus lainnya untuk menghasilkan output tertentu. Kombinasi escape character dengan karakter khusus lainnya disebut escape sequence.

Contoh beberapa kombinasi escape sequence di Python:

<br>

Escape Sequence    | Deskripsi
-------------------|------------------
\ + '              | Menambahkan tanda petik tunggal dalam teks (sudah dicoba)
\ + "              | Menambahkan tanda petik tunggal dalam teks (sudah dicoba)
\ + n              | Menambahkan baris baru dalam teks
\ + t              | Menambahkan tab horizontal dalam teks
\ + \              | Menambahkan backslash dalam teks

<br>


Tulis dan jalankkan code berikut kemudian analisis outputnya:

n.PNG

```python
print("\nFilosofi\nIlmu\nPadi")
```

t.PNG

```python
print("kian\tberisi\tkian\tmerunduk\t")
```

```python
print('nama : Ellsa')
print('tempat tinggal : Sidoarjo')
```

```python
print('nama\t\t : Ellsa')
print('tempat tinggal\t : Sidoarjo')
```

Escape sequence `\\` digunakan untuk menegatifkan efek escape character sehingga hanya dianggap sebagai backslash biasa.

Coba tulis dan jalankan dua code berikut, kemudian analisis output keduanya:


doublebackslash.PNG

```python
print('\tarik\non-tunai')
```

```python
print('\\tarik\\non-tunai')
```

Tentunya kita juga bisa mengkombinasikan escape sequence dengan escape sequence lainnya.

nt.PNG

```python
print("\nFilosofi \"Ilmu Padi\":\nkian\tberisi\nkian\tmerunduk\t")
```
